```python
variable = df['ApplicantIncome']
```

The central tendency of any two continuos variables

```python
import pandas as pd
```

```python
df = pd.read_csv('Loan_Approval_Data.csv')
```

```python
mean_var1 = df['ApplicantIncome'].mean()
mean_var2 = df['LoanAmount'].mean()
```

```python
print(f"The mean of continuous variable one, Application Income is, {mean_var1:.4f}")
print(f"The mean of continuous variable one, Loan Amount is, {mean_var2:.4f}")
```

```python
#CALCULATING THE MEDIAN
Median_var1 = df['ApplicantIncome'].median()
Median_var2 = df['LoanAmount'].median()
print(f"The median of continuous variable one, Application Income is, {Median_var1:.4f}")
print(f"The median of continuous variable one, Loan Amount is, {Median_var2:.4f}")

```

```python
#CALCULATING THE Mode
Mode_var1 = df['ApplicantIncome'].value_counts().idxmax()
Mode_var2 = df['LoanAmount'].mode()
print(f"The mode of continuous variable one, Application Income is, {Mode_var1:}")
print(f"The mode of continuous variable one, Loan Amount is, {Mode_var2:}")
```

```python
#comparison of central tendencies
if mean_var1 > mean_var2:
    print("The mean of Application Income is greater than the mean of Loan Amount ")
elif mean_var1 < mean_var2:
    print("The mean of Application Income is less than the mean of Loan Amount ")
else:
    print("The mean of Application Income is equal to the mean of Loan Amount" )
    
```

```python
if Median_var1 > Median_var2:
    print("The median of Application Income is greater than the median of Loan Amount ")
elif Median_var1 < Median_var2:
    print("The median of Application Income is less than the median of Loan Amount ")
else:
    print("The median of Application Income is equal to the median of Loan Amount" )
    
```

```python
if (Mode_var1 > Mode_var2).any():
    print("The mode of Application Income is greater than the mode of Loan Amount ")
elif Mode_var1 < Mode_var2:
    print("The mode of Application Income is less than the mode of Loan Amount ")
else:
    print("The mode of Application Income is equal to the mode of Loan Amount" )
    
```

The relationship between the variables 'GENDER' and 'Loan_Status

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Loan_Approval_Data.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender_Male'])
df['Loan_Status'] = le.fit_transform(df['Loan_Status'])
```

```python
var1 = df['Gender_Male']
var2 = df['Loan_Status']
```

```python
df = pd.get_dummies(df, columns = ['Gender'])
```

```python
correlation_coefficient = df['Gender_Male'].corr(df['Loan_Status'])
```

```python
print(f"The correlation coeficcient between gender and loan status is, {correlation_coefficient}")
```

```python
plt.scatter(var1, var2)
plt.ylabel('var1')
plt.xlabel("var2")
plt.title("Relationship between Gender and Loan Status")
plt.show()
```
