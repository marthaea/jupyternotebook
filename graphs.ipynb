NO4 <br/>A generated graph to show variablity or  dispersion on one continuous variable and one categorical variable

```python
Profit_x = df['Profit']
age_group_y = df['Age_Group']
```

```python
plt.scatter(Profit_x,age_group_y)
plt.xlabel('Profit')
plt.ylabel('Age Group')
plt.title('Line Plot showing variability or dispersion between one continuous variable and one categoric variable(Profit and Age)')
plt.show()
```

NO5<br/> Visuals showing the relationship between two continuous variables

```python
Profit_x = df['Profit']
Cost_y = df['Cost']
```

```python
plt.plot(Profit_x,Cost_y)
plt.xlabel('Profit')
plt.ylabel('Cost')
plt.title('Line Plot showing the relationship between two continuous variables(Profit and Cost)')
plt.show()
```

NO6 <br/>Relationship between two categorical values

```python
Product_x = df['Product']
age_group_y = df['Age_Group']
```

```python
plt.plot(Product_x,age_group_y)
plt.xlabel('Product')
plt.ylabel('Age Group')
plt.title('Line Plot showing the relationship between two categorical variables(Product and Age Group)')
plt.show()
```

NO7 <br/> Relationship between one continuous and one categorical value

```python
Profit_x = df['Profit']
age_group_y = df['Age_Group']
```

```python
plt.bar(Profit_x,age_group_y)
plt.xlabel('Product')
plt.ylabel('Age Group')
plt.title('Line Plot showing the relationship between one categorical and one continuous variable(Profit and Age Group)')
plt.show()
```
