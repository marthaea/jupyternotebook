```python
import pandas as pd
```

```python
df = pd.read_csv('Cancer_data(1).csv')
```

```python
my_variable = df['perimeter_mean']
```

```python
#calculating the original sample sizes
n = len(df)
print(n)
```

```python
df.shape
```

```python
import numpy as np
```

```python
#calculating the correlations between the observations
rho = np.var(df['perimeter_mean'])
```

```python
#calculating the effective sample size
n_eff = n / (1+ (n-1) * rho)
```

```python
print(f'The effective sample size of the perimeter mean is approximately {n_eff:.2f}')
```

Discussion <br>The effective size has been calculated by first extracting the variable, 'perimeter_mean' from our dataset and calculating the original size of the entire dataset, calculating the correlations between the observations and finally calculating the effective sample size 

NO5 <br> The marginal probability of the perimeter_mean

```python
import numpy as np
```

```python
my_variable = df['perimeter_mean']
```

```python
hist, bins = np.histogram(my_variable, bins=50, density= True )
```

```python
marginal_probability = hist/ np.sum(hist)
print(marginal_probability)
```

```python
#ploting the maginal probability
import matplotlib.pyplot as plt

plt.plot(bins[:-1], marginal_probability)
plt.xlabel('Value')
plt.ylabel('Marginal Probability')
plt.title('Marginal Probability of the perimeter_mean')
plt.show
```

NO6 <br> The probability of a tumour having a perimeter_mean less than 20

```python
#standard deviation of the perimeter_mean
stdd = df['perimeter_mean'].std()
```

```python
from scipy.stats import norm
import pandas as pd
```

```python
perimeter_values = df['perimeter_mean'] 
```

```python
if perimeter_values.dtype.kind in ['i', 'f']:
    mean_perimeter =perimeter_values.mean()
    std_perimeter = perimeter_values.std()
    
    probability =norm.cdf(20, loc=mean_perimeter, scale=std_perimeter)
    
    print(f"The probability that a tumour has perimeter mean less than 20 is, {probability:.4f}")
else:
    print("Error: perimeter values must be numeric")
```
