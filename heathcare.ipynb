```python
#The central tendency and distribution of any two continuous variables.
#The continuous variables used: Room Number and Billing amount.

#Extracting the variables from the dataset
first_continuous_variable = df['Billing Amount']
second_continuous_variable = df['Room Number']

#calculating the measeres of cental tendency
meanA = first_continuous_variable.mean()
medianA = first_continuous_variable.median()
modeA = first_continuous_variable.mode()
stdA = first_continuous_variable.std()
#Printing out the result for thr first continuous variable
print(f"The central tendency of the first continuous variable(Billing Amount) in terms of mean, median, mode and standard deviation respectinely is, {meanA}, {medianA}, {modeA}, {stdA}")
```

```python
#calculating the measeres of cental tendency
meanA = second_continuous_variable.mean()
medianA = second_continuous_variable.median()
modeA = second_continuous_variable.mode()
stdA = second_continuous_variable.std()
#Printing out the result for the second continuous variable
print(f"The central tendency of the second continuous variable(Room Number) in terms of mean, median, mode and standard deviation respectinely is, {meanA}, {medianA}, {modeA}, {stdA}")
```

**Marginal probability of any two categorical variables**

```python
#showing the variables we have
df.head(1)
```

```python
#The two categorical variables chosen are, gender and Doctor
#Marginal probability of gender
gender_counts = df['Gender'].value_counts()
total_count = len(df)
gender_probabilities = gender_counts / total_count
gender_probabilities
print(f"The marginal probability of gender is, {gender_probabilities}")
#The marginal probability of doctor

doctor_counts = (df['Doctor'].value_counts())
total_count = len(df)
doctor_probabilities = doctor_counts / total_count
doctor_probabilities
print(f"The marginal probability of Doctor is, {doctor_probabilities}")
```

**Explanation**
- Marginal probability is the probability of a variable existing independent of other variables.
- The marginal probabilities of gender and doctor are approximately 0.5 and 0.8 respectively, which means that the dataset is evenly distributed between females and males.
- From the calculated marginal probabilities, we can see that the probability of being a female is approximately 0.5, and the probability of being a male is also approximately 0.5. 
- The probability of being a doctor is approximately 0.7
